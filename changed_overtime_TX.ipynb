```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/drive/MyDrive/final/Group 3/paper_counts.csv'
data = pd.read_csv(file_path)

# Filter and aggregate data for the years 1799 to 1904
filtered_data = data[(data['Year'] >= 1799) & (data['Year'] <= 1904)]
aggregated_data = filtered_data.groupby('Year', as_index=False).sum()

# Plotting the data
plt.figure(figsize=(12, 6))
plt.plot(aggregated_data['Year'], aggregated_data['Count'], marker='o', linestyle='-', linewidth=2)
plt.title('Article Counts from 1799 to 1904', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Article Count', fontsize=14)
plt.grid(True)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

# Summary of the data
print("Data Summary:")
print(f"Year Range: {filtered_data['Year'].min()} to {filtered_data['Year'].max()}")
print(f"Total Articles: {filtered_data['Count'].sum()}")
```

```python
#ear range: 1799 to 1904.
# Number of articles Trends: The number of articles is relatively stable in most years, but there are significant fluctuations in some years.
# Data characteristics:
# The data are distributed in a discrete point pattern, and the number of articles is particularly concentrated.
# This may reflect fluctuations in scientific research activity at the time or incomplete collection of published data.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
articles_df = pd.read_csv('/content/drive/MyDrive/final/Group 3/articles.leptospirosis.csv')
paper_counts_df = pd.read_csv('/content/drive/MyDrive/final/Group 3/paper_counts.csv')

# Process articles data
articles_yearly_count = articles_df.groupby('Year').size().reset_index(name='Articles')
articles_yearly_count.rename(columns={'Year': 'Year', 'Articles': 'Articles'}, inplace=True)

# Process paper counts data
paper_counts_df.rename(columns={'Year': 'Year', 'Count': 'All_Papers'}, inplace=True)

# Merge the datasets
merged_data = pd.merge(articles_yearly_count, paper_counts_df, on='Year', how='outer').fillna(0)
merged_data['Year'] = merged_data['Year'].astype(int)  # Ensure Year is integer
merged_data.sort_values('Year', inplace=True)

# Plot with dual Y-axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot total papers on the first Y-axis
ax1.plot(
    merged_data['Year'],
    merged_data['All_Papers'],
    label='Total Papers',
    color='blue',
    marker='o',
    linewidth=2.5  # Increase line width
)
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Papers', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(axis='y', linestyle='--', linewidth=0.5)

# Plot leptospirosis articles on the second Y-axis
ax2 = ax1.twinx()
ax2.plot(
    merged_data['Year'],
    merged_data['Articles'],
    label='Leptospirosis Articles',
    color='orange',
    marker='x',
    linewidth=2.5  # Increase line width
)
ax2.set_ylabel('Leptospirosis Articles', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Title and layout
plt.title('Yearly Count of Papers: Leptospirosis vs Total (Dual Y-Axis)')
fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
articles_df = pd.read_csv('/content/drive/MyDrive/final/Group 3/articles.leptospirosis.csv')  # 
paper_counts_df = pd.read_csv('/content/drive/MyDrive/final/Group 3/paper_counts.csv')

#  articles 
articles_count_per_year = articles_df.groupby('Year').size()

#  paper_counts 
paper_counts_per_year = paper_counts_df.set_index('Year')['Count']

# 
years = range(2000, 2025)  #  2000  2024 
articles_count_aligned = articles_count_per_year.reindex(years, fill_value=0)
paper_counts_aligned = paper_counts_per_year.reindex(years, fill_value=0)

# Y
fig, ax1 = plt.subplots(figsize=(12, 6))

# Yarticles 
ax1.plot(years, articles_count_aligned, 'orange', label='Articles Count')
ax1.set_xlabel('Year')
ax1.set_ylabel('Articles Count (Orange)', color='orange')
ax1.tick_params(axis='y', labelcolor='orange')

# Ypaper_counts 
ax2 = ax1.twinx()
ax2.plot(years, paper_counts_aligned, 'b-', label='Paper Counts')
ax2.set_ylabel('Paper Counts (Blue)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# 
xticks = list(years)
ax1.set_xticks(xticks)
ax1.set_xticklabels(xticks)
for i, label in enumerate(ax1.get_xticklabels()):
    year = xticks[i]
    label.set_color('blue' if year % 2 == 0 else 'green')

# 
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))
plt.title('Articles Count vs Paper Counts (2000-2024)')
plt.grid(True)
plt.tight_layout()  # 
plt.show()
```

```python
import pandas as pd

# 
articles_df = pd.read_csv('/content/drive/MyDrive/final/Group 3/articles.leptospirosis.csv')

# 2008
filtered_data = articles_df[(articles_df['Year'] >= 2000) & (articles_df['Year'] <= 2024)]
year_2008_data = filtered_data[filtered_data['Year'] == 2008]

# 
leptospirosis_2008_data = year_2008_data[year_2008_data['Title'].str.contains('leptospirosis', case=False, na=False)]

# 
author_counts = leptospirosis_2008_data.groupby(['FirstAuthorForename', 'FirstAuthorLastname']).size().sort_values(ascending=False)

# 
categories = {
    'Epidemiology': ['epidemiology', 'outbreak', 'globalization', 'incidence', 'trend'],
    'Mechanism': ['renal', 'dysfunction', 'pathogenesis', 'magnesium', 'tubular'],
    'Diagnosis': ['molecular', 'detection', 'diagnosis', 'test'],
    'Complications': ['fatal', 'lethal', 'severe', 'pulmonary', 'failure'],
    'Risk Factors': ['risk', 'factors', 'associated', 'conditions']
}

# 
category_counts = {key: 0 for key in categories.keys()}
for title in leptospirosis_2008_data['Title']:
    for category, keywords in categories.items():
        if any(keyword.lower() in title.lower() for keyword in keywords):
            category_counts[category] += 1

# 
max_category = max(category_counts, key=category_counts.get)
max_percentage = (category_counts[max_category] / len(leptospirosis_2008_data)) * 100

# 
author_counts.head(), category_counts, max_category, max_percentage
```
